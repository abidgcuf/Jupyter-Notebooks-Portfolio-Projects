```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\hp21\Desktop\DATA ANALYSIS\Python Jupyter Notebook\Data\csv files\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
df.plot(kind = 'line',subplots = True)
```

```python
df.plot(kind = 'line',stacked = True,title = 'Ice Cream Rating', xlabel = 'Daily Ratings',ylabel = 'Scores')
```
